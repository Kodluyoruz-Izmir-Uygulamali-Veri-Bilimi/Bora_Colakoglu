# Missing Value Doldurma

Bundan sonraki soruları <b>train_missing.csv</b> datasını kullanarak cevaplayınız.

```python
df_titan_missing = pd.read_csv("train_missing.csv")
display(df_titan_missing)
df_titan_missing.isnull().sum()
```

# Soru 8

Age, Embarked ve Fare değişkenlerini mean/median/mode değerlerini kullanarak değişken tipine uygun olacak şekilde doldurunuz.

```python
df_titan_missing.info()
```

```python
embarked_mode = df_titan_missing.Embarked.mode()[0]
```

```python
# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız
df_titan_missing.Age = df_titan_missing.Age.fillna(df_titan_missing.Age.median())
df_titan_missing.Embarked = df_titan_missing.Embarked.fillna(embarked_mode)
df_titan_missing.Fare = df_titan_missing.Fare.fillna(df_titan_missing.Fare.mean())
```

```python
df_titan_missing.Age.isnull().sum()
```

```python
df_titan_missing.Embarked.isnull().sum()
```

```python
df_titan_missing.Fare.isnull().sum()
```

# Soru 9

SibSp değişkeni hangi yöntemler ile doldurulabilir, fikir yürütün. Uygun bir yöntem ile doldurun. (SibSp değişkeni kişilerin gemide bulunan kardeş/eş sayısını belirtiyor.)

```python
# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız
#benzer sosyo-ekonomik seviyede kisilerin benzer sayida yakini gemide olacaktir
#bu duruda knn algoritmasi kullanmak mantikli olur
```

```python
from sklearn.impute import KNNImputer
```

```python
df_titan_missing_2 = df_titan_missing.copy()
df_titan_missing_2.set_index("PassengerId", inplace = True)
```

```python
for col in ["Sex", "Ticket","Cabin","Embarked"]:
    df_titan_missing_2[col] = df_titan_missing_2[col].astype('category')
    df_titan_missing_2[col] = df_titan_missing_2[col].cat.codes
```

```python
X = df_titan_missing_2[df_titan_missing_2.columns.difference(["Name"])]
X
```

```python
imputer = KNNImputer(n_neighbors=5, weights="distance")
```

```python
df_titan_missing_2[df_titan_missing_2.columns.difference(["Name"])] = imputer.fit_transform(X)
```
